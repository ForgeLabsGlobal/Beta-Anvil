```python
#initializing variables
math_score_1 = data["math score"][0]
math_score_2 = data["math score"][1]
math_score_3 = data["math score"][2]
#displaying first 3 math scores from variables 
print("First student's math score = ", math_score_1)
print("Second student's math score = ", math_score_2)
print("Third student's math score = ", math_score_3)
#calculating total score
total_score = math_score_1 + math_score_2 + math_score_3
#display total
print("First Three students' total score = ", total_score)
```

```python
#initializing list
firstfivescores = (data["math score"][:5]).tolist()
#display first five scores
print(firstfivescores)
```

```python
#declaring var list
passed = []
#init var math_scores
math_scores = (data["math score"]).tolist()
#iterate through list for scores greater than or equal to 50
for score in math_scores:
    if score >= 50:
        passed.append(score)

#display output
print(len(passed))
```

```python
#init. var type list 
FirstFiveScores = data["math score"][:5].tolist()
#checking condition
for score in FirstFiveScores:
    if score >= 50:
        print("Student", FirstFiveScores.index(score) + 1 , "Pass")
    else:
        print("Student", FirstFiveScores.index(score) + 1 , "Fail")
```
