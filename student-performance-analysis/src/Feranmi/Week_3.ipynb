```python
import pandas as pd

#load students performance dataset
data = pd.read_csv(r"C:\Users\USER\Desktop\Forgelabs\res\data\StudentsPerformance.csv")

#initializing var math_scores to store list of candidates maths scores
math_scores = (data["math score"]).tolist()
#function to find students that passed
def count_passed():
    passed = 0 #counter
    
    #iterate through list for scores greater than or equal to 50
    for score in math_scores:
        if score >= 50:
            passed += 1  #updating number of candidates that passed
            
    #return updated value
    return passed
#display output
count_passed()
```
