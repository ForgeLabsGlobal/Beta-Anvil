```python
student1 = data["math score"][0]
student2 = data["math score"][1]
student3 = data["math score"][2]

total_score = student1 + student2 + student3

print("Total of first 3 students:", total_score)
```

```python
math_scores = list(data["math score"])

pass_count = 0

for score in math_scores:
    if score >= 50:
        pass_count += 1

print("Number of students that passed:", pass_count)
```

And also to learn the Conditional counting logic

```python
first_five_scores = list(data["math score"][:5])

for score in first_five_scores:
    if score >= 50:
        print("Pass")
    else:
        print("Fail")
```

this concept has just thought me how to distinguish and print based on scores if the student passed the course of failed the course

```python
first_ten_scores = list(data["math score"][:10])

pass_count = 0

for score in first_ten_scores:
    if score >= 50:
        pass_count += 1

print("Number of students that passed (first 10):", pass_count)
```
