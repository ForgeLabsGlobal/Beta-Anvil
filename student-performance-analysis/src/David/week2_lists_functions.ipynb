## What is a List?

A list allows us to store **multiple values in one variable**.

Example:

If we want to store several math scores, we can place them inside a list.

Lists are useful because datasets often contain hundreds of values.

```python
scores = [72, 69, 90, 47, 76]
```

In this example:

scores[0] gives the first value  
scores[1] gives the second value

Lists start counting from **0**.

Instead of typing scores manually, we can extract values from the dataset.

Your goal here is to create a list containing **math scores from the first 10 students**.

Steps to think about:

1. Locate the math score column.
2. Select the first 10 values.
3. Convert them into a list.

```python
# Create a list of the first 10 math scores

math_scores = data["math score"][0:10].to_list()

math_scores
```

## Functions

A function allows us to group logic together so we can reuse it.

Think of a function like a **machine**:

Input → Function → Output

Example:

Input: scores  
Function: calculate average  
Output: average score

```python
def example_function(value):
    result = value + 10
    return result
```

Important parts:

• def creates the function
• inputs go inside parentheses
• return sends the result back

Try creating a simple function that adds two numbers.

This is just practice to help you understand the structure of functions.

```python
def add_numbers(a, b):
    
    # combine the two numbers
    sum = a + b
    
    return sum
```

```python
add_numbers(4, 6)
```

### Task 2 —Average of Scores
Using the math_scores list you created earlier, write a function that calculates the **average score**.

Hints:

1. Add all values together.
2. Divide by the number of values in the list.

```python
def avg_math_score(list):
    sum = 0
    for i in list:
        sum += i
    average = sum / len(list)
    
    return average

avg_math_score(math_scores)
```

### Task 3 — Count Passing Students
A passing math score is **50 or higher**.

Create a function that counts how many students passed.

Hints:

• use a loop to check each score  
• use a counter variable  
• increase the counter when a score is ≥ 50

```python
def passed(score_list):
    counter = 0
    
    for score in score_list:
        if score >= 50:
            counter += 1
        
    return counter

passed(math_scores)
```

### Task 1 - Gender Analysis
Create a function that investigates **math performance by gender**.

Hints:

1. Filter the dataset for a specific gender.
2. Extract the math scores.
3. Store them in a list.
4. Compute the average.

```python

def gender_avg_score(gender, data):
    gender_list = data["gender"].to_list()
    score_list = data["math score"].to_list()
    
    scores = []
    sum = 0
    
    if gender == "male":
        for i in gender_list:
            if i == "male":
                scores.append(score_list[gender_list.index(i)])
    else:
        for i in gender_list:
            if i == "female":
                scores.append(score_list[gender_list.index(i)])
            
                
        
    for i in scores:
        sum += i
    average = sum / len(scores)
    
    return average
   
    
print(gender_avg_score("male", data))

print(gender_avg_score("female", data))
```
