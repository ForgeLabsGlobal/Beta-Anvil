## What Is a List?

A list stores multiple values in one place.

Example:
scores = [50, 60, 70]


```python
scores = [80, 90, 70]
print(scores[0])  # First item
print(scores[1:3])  # Slice: second and third items

```

Hint: Use slicing

## How to Access Values in a List or Column

A list or a column in Python is like a row of boxes.

- The first item is at index 0
- The second item is at index 1
- And so on…

Example:

scores = [80, 70, 90]
scores[0]  # This gets 80
scores[1]  # This gets 70

You can also get multiple values using slicing:

scores[0:2]  # Gets the first two items: [80, 70]

## What Is a Conditional?

A conditional allows Python to make decisions.

Example:
If score is above 50 → pass

```python
score = 65

if score >= 50:
    print("Pass")
else:
    print("Fail")
```
