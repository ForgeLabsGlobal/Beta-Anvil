## Understanding Loops

A loop allows us to repeat an action many times.

Example:

If we have a list of scores, we can inspect each score one by one.

```python
scores = [72, 69, 90, 47]

for score in scores:
    print(score)
```

What this code does:

• `for score in scores` means:  
  take each value in the list one at a time.

• `print(score)` displays each value.

The loop continues until every value has been checked.
