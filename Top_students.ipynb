```python
import pandas as pd
import numpy as np
```

```python
 # Function that finds the top-performing students in a subject.

def get_top_scores(student_data, subject_column, num_top_scores=3):
    """
    Finds the highest scores in a specific subject.
    """
    # 1. Convert column data into a list
    scores = list(student_data[subject_column])

    # 2. Sort the list in descending order (highest to lowest)
    scores.sort(reverse=True)

    # 3. Return only the highest values using list slicing
    top_scores = scores[:num_top_scores]

    return top_scores
```
